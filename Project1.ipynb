```python
#importing dependencies #span mail project
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


path= r'E:\PythonPracticenew\datasets\mail_data.csv'
df = pd.read_csv(path)
```

```python
#checking whether there are null values and changing null values with null strings
df.where((pd.notnull(df)),'')
```

```python
#printing the first 5 rows of the dataset
df.head()
```

```python
#checking the number of rows and columns in the dataframe
df.shape
```

```python
#replace the text value in Categorical Variable
df.replace({'Category':{'ham':1,'spam':0}},inplace=True)
df.head()
```

```python
#seperating target and feature variables
X = df['Message']
Y = df['Category']
```

```python
#splitting the data in train and test split 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.2,stratify=Y,random_state=3)
```

```python
#transform the test data to feature vecors that can be used as input to Logistic Regression Model
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase = True)

X_train_features = feature_extraction.fit_transform(X_train)
#dont use fit for the second one 
X_test_features = feature_extraction.transform(X_test)

#convert Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
#modelling the Logistic Regression model
model = LogisticRegression().fit(X_train_features,Y_train)
model
```

```python
#model evulation on train data
X_train_predict = model.predict(X_train_features)
predict_accuracy = accuracy_score(X_train_predict,Y_train)
print(predict_accuracy)
```

```python
#model evulation on test data
X_test_predict = model.predict(X_test_features)
predict_accuracy_test = accuracy_score(X_test_predict,Y_test)
print(predict_accuracy_test)
```
